```python
import os, sys, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = r"C:\Users\ADMIN\Desktop\NDM_Project"
sys.path.append(os.path.join(ROOT, "src"))

from dataio import (
    read_labels_csv, make_virtual_records, split_records_stratified,
    save_splits_json, load_splits_json, summarize_splits, iter_segments_from_record
)

print("ROOT =", ROOT)
```

```python
labels = read_labels_csv(ROOT)
display(labels)

FS = 48000
CHUNK_SECONDS = 1.0        # "file ảo" = chunk 1s, không chồng lấp
SEED = 42

records = make_virtual_records(ROOT, labels, fs=FS, chunk_seconds=CHUNK_SECONDS, drop_last=True)
print("Num virtual records =", len(records))

splits = split_records_stratified(records, seed=SEED, train_ratio=0.8, val_ratio=0.1, test_ratio=0.1)
print({k: len(v) for k, v in splits.items()})

out = save_splits_json(
    ROOT,
    r"data\splits\cwru_splits.json",
    records,
    splits,
    meta={"fs": FS, "chunk_seconds": CHUNK_SECONDS, "seed": SEED, "note": "Split by non-overlapping chunks (anti-leak) before segmentation"}
)
print("Saved:", out)
```

```python
records_map, splits2, meta = load_splits_json(ROOT, r"data\splits\cwru_splits.json")
print("meta =", meta)

summary = summarize_splits(records_map, splits2)
print("Counts per split per label:")
display(pd.DataFrame(summary).fillna(0).astype(int))

# Demo: lấy 1 record bất kỳ trong train để segment
rid0 = splits2["train"][0]
rec0 = records_map[rid0]
print("Example record_id =", rid0)
print("rec =", rec0)

SEG_LEN = 4096     # ~0.085s ở 48kHz
HOP_LEN = 4096     # không overlap (đơn giản, chống leak rõ ràng)

segs = list(iter_segments_from_record(ROOT, rec0, seg_len=SEG_LEN, hop_len=HOP_LEN, normalize=True))
print("num segments from this record =", len(segs))
print("first seg mean/std =", float(np.mean(segs[0])), float(np.std(segs[0])))

plt.figure()
plt.plot(segs[0][:800])
plt.title("First normalized segment (first 800 samples)")
plt.show()
```
