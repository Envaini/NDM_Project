```python
import os, sys, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = r"C:\Users\ADMIN\Desktop\NDM_Project"
sys.path.append(os.path.join(ROOT, "src"))

from dataio import load_splits_json, iter_segments_from_record
from features_prml import extract_features

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

records_map, splits, meta = load_splits_json(ROOT, r"data\splits\cwru_splits.json")
FS = int(meta["fs"])
print("FS =", FS)
print({k: len(v) for k, v in splits.items()})
```

```python
SEG_LEN = 4096
HOP_LEN = 4096
NORMALIZE_SEG = False  # baseline feature nên lấy từ raw segment; scaler sẽ xử lý

def build_Xy(split_name: str):
    X_list, y_list = [], []
    for rid in splits[split_name]:
        rec = records_map[rid]
        for seg in iter_segments_from_record(
            ROOT, rec,
            seg_len=SEG_LEN, hop_len=HOP_LEN,
            normalize=NORMALIZE_SEG
        ):
            feat = extract_features(seg, fs=FS)
            X_list.append(feat)
            y_list.append(rec["label"])
    X = np.vstack(X_list).astype(np.float32)
    y = np.array(y_list)
    return X, y

X_train, y_train = build_Xy("train")
X_val, y_val = build_Xy("val")
X_test, y_test = build_Xy("test")

print("Shapes:", X_train.shape, X_val.shape, X_test.shape)

le = LabelEncoder()
ytr = le.fit_transform(y_train)
yva = le.transform(y_val)
yte = le.transform(y_test)

print("Classes:", list(le.classes_))
```

```python
def eval_model(model, Xtr, ytr, Xva, yva):
    model.fit(Xtr, ytr)
    pva = model.predict(Xva)
    acc = accuracy_score(yva, pva)
    f1 = f1_score(yva, pva, average="macro")
    return acc, f1

results = []

# 1) Logistic Regression (multiclass)
for C in [0.01, 0.1, 1.0, 10.0, 100.0]:
    clf = Pipeline([
        ("scaler", StandardScaler()),
        ("lr", LogisticRegression(C=C, max_iter=5000, multi_class="auto"))
    ])
    acc, f1 = eval_model(clf, X_train, ytr, X_val, yva)
    results.append(("LogReg", {"C": C}, acc, f1))

# 2) SVM (RBF)
for C in [0.1, 1.0, 10.0]:
    for gamma in ["scale", 0.01, 0.1, 1.0]:
        clf = Pipeline([
            ("scaler", StandardScaler()),
            ("svm", SVC(C=C, gamma=gamma, kernel="rbf"))
        ])
        acc, f1 = eval_model(clf, X_train, ytr, X_val, yva)
        results.append(("SVM_RBF", {"C": C, "gamma": gamma}, acc, f1))

# 3) Random Forest (không cần scaler)
for n_estimators in [200, 500]:
    for max_depth in [None, 10, 20]:
        clf = RandomForestClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            random_state=42,
            n_jobs=-1
        )
        acc, f1 = eval_model(clf, X_train, ytr, X_val, yva)
        results.append(("RF", {"n_estimators": n_estimators, "max_depth": max_depth}, acc, f1))

df = pd.DataFrame(results, columns=["model", "params", "val_acc", "val_macro_f1"])
df = df.sort_values(["val_macro_f1", "val_acc"], ascending=False).reset_index(drop=True)
df.head(15)
```

```python
best = df.iloc[0]
best_model = best["model"]
best_params = best["params"]
print("BEST (by val):", best_model, best_params, "val_acc=", best["val_acc"], "val_f1=", best["val_macro_f1"])

def build_best_clf(model_name, params):
    if model_name == "LogReg":
        return Pipeline([
            ("scaler", StandardScaler()),
            ("lr", LogisticRegression(C=params["C"], max_iter=5000, multi_class="auto"))
        ])
    if model_name == "SVM_RBF":
        return Pipeline([
            ("scaler", StandardScaler()),
            ("svm", SVC(C=params["C"], gamma=params["gamma"], kernel="rbf"))
        ])
    if model_name == "RF":
        return RandomForestClassifier(
            n_estimators=params["n_estimators"],
            max_depth=params["max_depth"],
            random_state=42,
            n_jobs=-1
        )
    raise ValueError("Unknown model_name")

clf = build_best_clf(best_model, best_params)
clf.fit(X_train, ytr)
pte = clf.predict(X_test)

test_acc = accuracy_score(yte, pte)
test_f1 = f1_score(yte, pte, average="macro")
print("TEST: acc=", test_acc, "macro_f1=", test_f1)

# save metrics csv (tất cả candidates + best test)
df_out = df.copy()
df_out["test_acc"] = np.nan
df_out["test_macro_f1"] = np.nan
df_out.loc[0, "test_acc"] = test_acc
df_out.loc[0, "test_macro_f1"] = test_f1

os.makedirs(os.path.join(ROOT, "results"), exist_ok=True)
csv_path = os.path.join(ROOT, "results", "baseline_prml_metrics.csv")
df_out.to_csv(csv_path, index=False)
print("Saved:", csv_path)

# confusion matrix for best model on test
cm = confusion_matrix(yte, pte, labels=np.arange(len(le.classes_)))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
fig, ax = plt.subplots(figsize=(8, 8))
disp.plot(ax=ax, xticks_rotation=45, colorbar=False)
ax.set_title(f"PRML baseline best: {best_model}")
plt.tight_layout()

png_path = os.path.join(ROOT, "results", "cm_prml.png")
plt.savefig(png_path, dpi=200)
plt.show()
print("Saved:", png_path)
```
